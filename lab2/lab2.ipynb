```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

В данном датасете нужно определить выдавать ли кредитную карточку человеку или нет на основе некоторых параметров. Делать это будем при помощи модели гауссовского наивного байесовского классификатора.

```python
data=pd.read_csv('Credit_card.csv')
ans=pd.read_csv('Credit_card_label.csv')
data=pd.merge(data,ans,on='Ind_ID')
data
```

Можно увидеть что в некоторых строках пропущены столбцы. Удалим их, а также сделаем из категориальных признаков числовые.

```python
data = data.drop(['Type_Occupation','Mobile_phone'], axis=1)
data = data.dropna(subset=['GENDER','Annual_income','Birthday_count'])
for column in data:
    if type(data[column][0]) is str:
        data[column] = pd.factorize(data[column])[0]
data
```

```python
sns.heatmap(
    round(
        abs(data.corr()),
        1,
    ),
    annot=True,
) 
```

На графике можно заметить что есть мультиколлинеарность. Для того что от неё избавиться удалим параметры 'Family_Members','Employed_days','Birthday_count','Car_Owner'

```python
data=data.drop(['Family_Members','Employed_days','Birthday_count','Car_Owner'],axis=1)
```

Разделим выборку на тренировочную и тестовую

```python
train_input, test_input, train_output, test_output = train_test_split(
    data.drop('label', axis=1),
    data["label"],
    test_size=0.2
) 
```

Построим модель гауссовского наивного байесовского классификатора(y = label, x = все остальные параметры).

```python
model = GaussianNB()
model.fit(train_input, train_output)
predictions = model.predict(test_input)
accuracy = metrics.accuracy_score(predictions, test_output)
print(f"Точность модели на тестовом участке = {accuracy}") 
```

Точность модели достаточно высокая.

Из данного датасета можно построить модель, которая будет определять выдать ли человеку кредитную карточку на основе некоторых параметров. Это поможет лучше определять не благонадежных клиентов и увеличит прибыль банковских учреждений.
